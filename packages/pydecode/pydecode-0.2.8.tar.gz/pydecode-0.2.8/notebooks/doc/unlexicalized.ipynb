```python
dep_parse = [-1, 2, 0, 2]
```

```python
def convert_to_matrix(dep_parse):
    n = len(dep_parse)
    matrix = np.zeros((n, n))
    for mod, head in enumerate(dep_parse):
        if mod == 0: continue
        matrix[mod, head] = 1
    return matrix
```

```python
convert_to_matrix(dep_parse)
```

```python
def make_index_set(n):
    return np.arange(n*n).reshape((n, n))
def make_items(items, n):
    return np.array(np.unravel_index(items, (n,n))).T

def make_label_set(n):
    return np.arange(n*n*n).reshape((n, n, n))
def make_labels(labels, n):
    return np.array(np.unravel_index(labels, (n,n,n))).T
```

```python
def xbar_cky(n):
    items = make_index_set(n)
    labels = make_label_set(n)
    chart = pydecode.ChartBuilder(items)

    # chart.init(np.diag(items))
    # C[i,i] = semiring.1

    for i in range(n):
        chart.init(items[i,i])


    for d in range(1, n):
        for i in range(n):
            k = i + d
            if k >= n: continue
            # set chart C[i,k] = \bigoplus_{j \in {i,k-1} }  C[i,j] \otimes C[j+1,k]
            chart.set(items[i, k],
                      [[items[i, j], items[j+1, k]]
                       for j in range(i, k)],
                      labels=[labels[i,j,k] 
                              for j in range(i, k)])
            # chart.set_t(items[i, k],
            #             items[i, i:k],
            #             items[i+1:k+1, k],
            #             labels=labels[i, i:k, k])


    return chart.finish()

            # for j in range(i, k):
            #     print (i,j,k)
```

```python
graph = xbar_cky(4)
pydecode.draw(graph, make_labels(graph.labeling, 4), make_items(graph.node_labeling, 4))
```

```python
weights = np.ones(len(graph.edges), dtype=np.int32)
inside = pydecode.inside(graph, weights, weight_type=pydecode.Counting)
inside
pydecode.draw(graph, make_labels(graph.labeling, 4), inside)
```
