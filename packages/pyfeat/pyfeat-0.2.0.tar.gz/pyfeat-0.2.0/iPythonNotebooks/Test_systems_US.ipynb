```python
#imports
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
# import API functions for Reader, MBAR, WHAM
from pyfeat import read_files, run_mbar, run_wham
```

## Test case: Symetric double well US

#### MBAR

```python
files = []
for i in xrange(23):
    files.append('data/US_SDW/replica'+str(i)+'.dat')
data_sdw_us = read_files(
        files,
        wham_file='data/US_SDW/us_wham.dat',
        verbose=False
    )
```

```python
mbar_sdw_us = run_mbar(
        data_sdw_us,
        maxiter=800,
        ftol=0.00001,
        verbose=False
    )
```

```python
exact_sdw_us = np.loadtxt("data/US_SDW/exact.dat")
plt.plot(exact_sdw_us[:,0], exact_sdw_us[:,1]/exact_sdw_us[:,1].sum(), '--o')
plt.plot(exact_sdw_us[:,0], mbar_sdw_us.pi_markov, '--^' )
plt.xlabel('distance x')
plt.ylabel('probability P(x)')
plt.semilogy()
```

```python
plt.plot(exact_sdw_us[:,0], -np.log(exact_sdw_us[:,1]/exact_sdw_us[:,1].sum()), '--o')
plt.plot(exact_sdw_us[:,0], mbar_sdw_us.fi_markov, '--^' )
plt.xlabel('distance x')
plt.ylabel ('Free energy f(x)')
```

#### WHAM

```python
wham_sdw_us = run_wham(
        data_sdw_us,
        maxiter=800,
        ftol=0.00001,
        verbose=False
    )
```

```python
plt.plot(exact_sdw_us[:,0], exact_sdw_us[:,1]/exact_sdw_us[:,1].sum(), '--o')
plt.plot(exact_sdw_us[:,0], wham_sdw_us.pi_markov, '--^' )

plt.xlabel('distance x')
plt.ylabel('probability P(x)')
```

```python
plt.plot(exact_sdw_us[:,0], -np.log(exact_sdw_us[:,1]/exact_sdw_us[:,1].sum()), '--o')
plt.plot(exact_sdw_us[:,0], wham_sdw_us.fi_markov, '--^' )
plt.xlabel('distance x')
plt.ylabel ('Free energy f(x)')
```
